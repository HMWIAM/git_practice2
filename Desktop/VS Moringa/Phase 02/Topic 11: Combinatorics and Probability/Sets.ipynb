# Introduction to sets

A set is defined as a well defined collections of distinct objects

```python
S = {1,2,3}
type(S)
```

#### Set Membership

If an element x belongs in set S, then\
If an element x does not belongs in set S, then

```python
print(1 in S)
print(1 not in S)
```

#### Working with multiple sets

```python
# Setting up example sets
A = {1,2,3}
B = {1,2,3}
C = {1,2}
```

```python
# B and A is a subset of A as it contains elements of A
print(C.issubset(A))
print(B.issubset(A))
#<= can be used to check if something is a subset
print(C <= A)
print(B <= A)
```

#### Proper subsets

A proper subset is defined as a set that contains elements of other sets, however, they do not contains ALL elements of that set

```python
# We're asking "is C a proper subset of A?"
print(C < A)
# True, because every element of C is in A, and A contains at least 1 element that C does not

# We're asking "is A a proper subset of B?"
print(A < B)
# False, because even though every element of A is in B,A and B have the exact same elements
```

A subset (<=) includes sets that are equal(have same elements) \
A proper subset (<) only includes sets that have fewer elements

#### Supersets

A superset is kust the inverse of a subset.\
In python, this can be represented using > for proper superset and >=(.issupperset method) for superset

```python
# We're asking "is A a superset of C?"
print(A >= C)
# We're asking "is A a proper superset of C?"
print(A > C)
# We're asking "is A a superset of B?"
print(A >= B)
# You can also use this method if you're only checking
# for a superset, not a proper superset
print(A.issuperset(C))
```

Absolute complement are all the elements that are not in the universal set. \

## The Inclusion Exclusion Principle
Recall from the previous lesson that:

$$\mid A \cup B\cup C\mid = \mid A \mid + \mid B \mid + \mid C \mid - \mid A \cap B \mid  -\mid A \cap C \mid - \mid B \cap C \mid  + \mid A \cap B \cap C \mid $$

Combining these main commands:

| Method        |	Equivalent |	Result |
| ------                    | ------       | ------    |
| a.union(b)                |	A $\mid$ B | new set with elements from both a and b
| a.intersection(b)         |	A & B      | new set with elements common to a and b

along with the `len(x)` function to get to the cardinality of a given x ("|x|").

What you'll do is translate the left hand side of the equation for the inclusion principle in the object `left_hand_eq`, and the right hand side in the object `right_hand_eq` and see if the results are the same.

```python
left_hand_eq = len(A | B | C)
print(left_hand_eq)  # 9 elements in the set
right_hand_eq = len(A) + len(B) +len(C) - len(A & B) - len(A & C) - len(B & C) + len(A&B&C)
print(right_hand_eq) # 9 elements in the set
left_hand_eq == right_hand_eq # Use a comparison operator to compare `left_hand_eq` and `right_hand_eq`. Needs to say "True".
```
